```python
import Majorana_main  as Maj
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def convert_to_parser(string):
    return string.split(' ')+['-wavefunction']
```

```python
oneline='-L=1 -mu=1 -lead_num=1 -lead_pos=LR -SE -x_max=1.2 -dissipation=0.00 -barrier_E=0.0'
```

```python
parser=Maj.argparse.ArgumentParser()
args=Maj.parse_arguments(parser,convert_to_parser(oneline))
```

```python
nw=Maj.Nanowire(args)
```

```python
trial=[0.024195,0.059068]
fig,axs=plt.subplots(1,2,figsize=(6.8*2,4))
results=[]
for y,ax in zip(trial,axs):
    result=nw.wavefunction(1.1,y)
    Maj.plot_wavefunction(result,args,fig,ax)
    results.append(result)
```
